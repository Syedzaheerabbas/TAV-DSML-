
### <font color='purple'>**Q1. Calculate the Probability of flood given that rainfall in June is greater than the median june rainfall value (threshold for heavy rainfall)**</font>



**Question Explanation**:  


Let A represents : Flood  
B represents: heavy rain in June  

We need to calculate $P(A | B)$ i.e. $\frac{P(A ∩ B)}{P(B)}$  

<br>  

  




**In this case:**  

By setting <font color='orange'>**normalize='index'**</font>,   

- the code calculates conditional probabilities within each row.   

- Each value in the table represents the probability of the corresponding event (FLOODS) given the value of 'JUN' in that row.   

The row sums up to 1, ensuring that it reflects the conditional probabilities.   

<br>

In summary,   

<font color='purple'>**setting normalize='index'** in pd.crosstab allows you to calculate and visualize conditional probabilities based on the specified row variable ('JUN' in this case)</font>,  

making it easier to assess the impact of one variable on another.

```python
pd.crosstab(index = df['JUN'],
            columns = df['FLOODS'],
            margins=True,
            normalize='index')
```



The values in the table represent the conditional probabilities, where each cell contains the probability of the corresponding outcome (FLOODS) given the condition in June (JUN).  


> <font color='purple'>Then the probability of flood occurring given that the heavy rainfall occured in June will be:</font>

- In the cell at row 1, column 1, the value <font color='orange'>**0.73333**</font> represents the conditional probability of flooding (FLOODS = YES) given that high rainfall occurred in June (JUN = 1).  

#### <font color='orange'>Conclusion:</font>


      So, there is 73.33% chance of Floods when there is a heavy rainfall in June

As we can see by calculating using formula also, we are getting the same answer as using directly conditional probability using normalize = 'index'

### <font color='purple'>**Q2. Given that there is a flooding, calculate the probability that heavy rainfall has occurred in July (more than threshold value)?**</font>

Here we want to find <font color='purple'>$P(July = 1 | Flood = YES)$</font>   

We are already aware of using formula based approach, so We will solve this using contingency table  

Before proceeding,  

> <font color='purple'>**Q. In this question, which string will be passed inside normalize=' ' attribute?  'index' or 'columns'**</font>    

In this question, we should normalize the contingency table along the columns   

- As we want to find the conditional probability of <font color='purple'>**high rainfall in July (JUL = 1) given that there was flooding (FLOODS = YES)**</font>,   

We want to see how the 'JUL' column behaves when there is flooding.



#### <font color='purple'>**Solution:**</font>

```python
pd.crosstab(index = df['JUL'],
            columns = df['FLOODS'],
            margins=True,
            normalize='columns')
```

#### <font color='orange'>**Conclusion:**</font>  

The probability that high rainfall occurred in July (JUL = 1) given flooding (FLOODS = YES) is **0.65**.    

- **This means that when there is flooding, there is a 65% chance of heavy rainfall in July.**

### <font color='purple'>**Q3.Calculate the probability of flood given that june and july rainfall was greater than their median rainfall value?**</font>  





### **Solution:**  

We want to find $P(Flood = Yes | \ june = 1 \ and\ Jul = 1)$  

Here, we can pass multiple columns in the **pd.crosstab()**  



```python
pd.crosstab(index = [df['JUN'], df['JUL']],
            columns = df['FLOODS'],
            margins=True,
            normalize='index')
```

### <font color='orange'>**Conclusion**</font>  

Frequency (JUN = 1, JUL = 1, FLOODS = YES ) = 0.9000000

There is **90%** chance of flood given that heavy rainfall in both june and july
