#Loops Problem Solving

```python
#HCF
n1=int(input())
n2=int(input())
for i in range(1,min(n1,n2)+1):
  if n1%i == 0 and n2%i == 0:
    hcf=i
print(hcf)
```

```python
#LCM
n1=int(input())
n2=int(input())
for i in range(min(n1,n2),n1*n2+1):
  if i%n1 == 0 and i%n2==0:
    print(i)
    break
```
