##While Loop

```python
# I want to print all the numbers from 0  to 7(include 7)
number = 0
while number <=7:
  print(number, end=' ')
  number += 1
```

```python
'''
Question 7 -
Print all the numbers from an input N to 1 in reverse order.
WITHOUT USING NEGATIVE JUMP!
N > 1 -> Given
'''
n=int(input())
for i in range(n+1):
  print(n-i,end=" ")
```
