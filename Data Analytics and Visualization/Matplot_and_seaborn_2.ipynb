##Univariate Analysis

```python
df.head()
```

**Categorical Colum**

```python
sns.countplot(data=df,x='Genre')
plt.xticks(rotation = 90)
plt.show()
```

**Numerical Column**


```python
sns.kdeplot(data=df,x='Global_Sales')
```

##Bi-variate Analysis

**Numerical - Numerical**

```python
sns.lineplot(data=df,x='Year',y='Global_Sales',label='Gloabl_Sales')
```

```python
sns.lineplot(data=df,x='Year',y='Global_Sales',label='Gloabl_Sales')
sns.lineplot(data=df,x='Year',y='NA_Sales',label='NA_Sales')
plt.legend(loc = 1,frameon= False,mode='expand')
```

```python
sns.scatterplot(data=df,x='Global_Sales',y='NA_Sales')
plt.title('Distribution of NA Sales vs Global Sales')
plt.show()
```

**Category - Category**

```python
df.head()
```

```python
top3_genre = df['Genre'].value_counts().index[:3]
top3_pub = df['Publisher'].value_counts().index[:3]
top3_plat= df['Platform'].value_counts().index[:3]
top3_data = df[(df['Genre'].isin(top3_genre)) & (df['Publisher'].isin(top3_pub)) & (df['Platform'].isin(top3_plat))]
top3_data

```

```python
sns.countplot(data = top3_data , x = 'Genre', hue = 'Publisher')
```

**Categorical - Numercal**

```python
sns.boxplot(data=top3_data,x='Genre',y='Global_Sales')
```

```python
sns.barplot(data=top3_data,x='Genre',y='Global_Sales', estimator = np.mean)
```

```python
sns.scatterplot(data=top3_data,x='Global_Sales',y='NA_Sales',hue = 'Genre')
plt.title('Distribution of NA Sales vs Global Sales')
plt.show()
```

##Multi- Variate Analysis

```python
sns.pairplot(data=top3_data)
```
