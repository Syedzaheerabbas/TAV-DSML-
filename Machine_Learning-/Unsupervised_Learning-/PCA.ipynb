```python
import numpy as np
import pandas as pd
```

```python
data = np.array([[3,7] ,
                [-4,-6] ,
                [1,-1] ,
                [7,8] ,
                [-4,-1] ,
                [-3,-7]])
```

```python
dataframe = pd.DataFrame(data , columns = ['x1', 'x2'])
```

```python
dataframe
```

Step 1 : Convert the data in Standard Normal format

`Standard Normal format: If it contins mean as 0 and Standard deviation as 1`.

```python
dataframe.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(dataframe)
```

```python
dataframe = pd.DataFrame(data_scaled, columns = ['x1', 'x2'])
```

```python
dataframe.describe()
```

Now we compute covarience Matrix of two features x1 & x2 cov(x1 , x2)

```python
#Approach 1
c1 = dataframe.x1
c2 = dataframe.x2
np.cov(c1 , c2)
```

```python
#Approach 2
covariancematrix = dataframe.T @ dataframe /5
covariancematrix
```

```python
eigenvalues, eigenvectors = np.linalg.eig(covariancematrix)
```

```python
## eagenvalues = fwatures
#eagenvalue represents the strength of the information given by the eagen vector
eigenvalues
```
