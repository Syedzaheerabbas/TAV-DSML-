```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (12,8)
```

```python
df = pd.read_csv('/content/AnomalyDetection.csv')
df.head()
```

```python
plt.scatter(df.Mileage, df.Price)
plt.yticks(np.arange(40,160,20))
plt.xticks(np.arange(0,160,10))
plt.xlabel('Mileage')
plt.ylabel('Price')
```

```python
from sklearn.ensemble import IsolationForest
```

```python
clf = IsolationForest(random_state=0, contamination = 0.08)
df['is_outlier_if'] = clf.fit_predict(df.iloc[:, :-1])
df['is_outlier_if']
```

```python
plt.scatter(df.Mileage, df.Price, s=40, c=df['is_outlier_if'])
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```

```python

scores = clf.decision_function(df.iloc[:, :-1])
print("Anomaly Scores:", scores)
```
