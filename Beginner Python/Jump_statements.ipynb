#Jump Satements

## Pass
The pass statement is used as a placeholder for future code

```python
for i in range(6):
  if i%2==0:
    pass
  else:
    print(i,' is odd')
```

## Continue

```python
for i in range(6):
  if i%2==0:
    continue
  else:
    print(i,' is odd')
```

## Break
'Break' in Python is a loop control statement.

```python
number = 1
while number <=5:
  if number ==3:
    break
  print(number)
  number +=1
```

## GCD or HCF

```python
a= 5
b= 15
hcf=1
for i in range(1,6):
  if 5%i==0 and 15%i==0:
    hcf = i
print(hcf)
```

```python
a= int(input())
b= int(input())
hcf=1
for i in range(1,min(a,b)+1):
  if a%i==0 and b%i==0:
    hcf = i
print(f"Hcf of {a} and {b} is ",hcf)
```
