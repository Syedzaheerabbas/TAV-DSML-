```python
l = [2, 0, "hello", None]


for element in l:
  try:
    print(f'Current element - {element}')
    result = 5/int(element)
    print(f'Result - {result}')

  except:
    print('An exception occured')

  print('-' * 25)

print('Execution done')
```

```python
l = [2, 0, "hello", None]

for element in l:
  try:
    print(f"Current Element - {element}")
    result = 5 / int(element)
    print(f"Result - {result}")
  except Exception as e:
    print(f"EXCEPTION - {e}")

  print("-" * 25)

print("EXECUTION DONE!")
```

```python
l = [2, 0, "hello", None]

for element in l:
  try:
    print(f"Current Element - {element}")
    result = 5 / int(element)
    print(f"Result - {result}")
  except ZeroDivisionError as z:
    print("YOU DIVIDED BY ZERO BY MISTAKE!")
    print("DO BETTER NEXT TIME!")
  except Exception as e:
    print(f"EXCEPTION - {e}")

  print("-" * 25)

print("EXECUTION DONE!")
```
