**What is File Handling?**

File handling in Python allows us to create, read, write, and modify files stored on disk.

**Files help us:**

Store data permanently

Read large data efficiently

Work with logs, reports, datasets, etc.

File Modes in Python
| Mode   | Meaning                 |
| ------ | ----------------------- |
| `"r"`  | Read file               |
| `"w"`  | Write file (overwrites) |
| `"a"`  | Append file             |
| `"r+"` | Read & Write            |


Reading from a File

```python
s = open('demo.txt' , mode = 'r')
print(s.read())
s.close()
```

Writing a file

```python
s = open('demo.txt' , mode = 'w')
s.write('Close')
s.close()
```

Appending into a file


```python
s = open('demo.txt' , mode = 'a')
s.write('Close bye')
s.close()
```

Reading and writing a file


```python
s = open('demo.txt' , mode = 'r+')
print(s.read())
s.write('okok')
s.close()
```

Write a program that writes a list of tuples, each containing a student's name and grade, to a text file. Each line in the file should contain the student's name and grade separated by a comma. Assume the file is named grades.txt.

```python
students = [('Alice', 'B') , ('Bob' , 'B') , ('Charlie' , 'C')]
```

```python
with open('demo.txt' , mode = 'w') as file:
  for name , grade in students:
    file.write(f'{name} , {grade}\n')
print("Student grades written to grades.txt successfully.")
```

Write a program that reads a CSV file containing product names and their prices, then calculates and prints the total cost of all products. Assume the file is named products.csv and has the format: ProductName,Price

```python
total_cost = 0

with open("Products.csv", "r") as file:

    next(file)

    for line in file:
        product, price = line.strip().split(",")
        total_cost += float(price)

print("Total cost of all products:", total_cost)
```

**Line-by-Line Explanation**

1️⃣ total_cost = 0

We create a variable total_cost to store the sum of all product prices.

Initialized to 0 because we haven’t added any prices yet.


2️⃣ with open("products.csv", "r") as file:

**open("products.csv", "r")** → Opens the file products.csv in read mode (r).

**with** → ensures the file is automatically closed after the block, even if an error occurs.

**as file** → gives us a variable file to work with the file content.


3️⃣ next(file)

Skips the first line of the CSV file, which is the header (ProductName,Price).

We don’t want to process the header as a product.


4️⃣ for line in file:

Loops through each remaining line in the file.

Example: "Laptop,50000\n" is one line.


5️⃣ product, price = line.strip().split(",")

**line.strip()** → removes leading/trailing spaces and newline \n.

**.split(",")** → splits the line at the comma, returning a list of two elements: [product_name, price].

**product, price** = → assigns the first element to product, the second to price
